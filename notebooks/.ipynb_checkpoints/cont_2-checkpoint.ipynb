# Contribution 2 Evidence (Ablation)

Component-wise ablation shows each guard element is necessary.


```python
import json
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().resolve()
for _ in range(4):
    if (PROJECT_ROOT / 'results').exists() and (PROJECT_ROOT / 'src').exists():
        break
    PROJECT_ROOT = PROJECT_ROOT.parent

RESULTS_ROOT = PROJECT_ROOT / 'results'
FIG_DIR = RESULTS_ROOT / 'figures'
TAB_DIR = RESULTS_ROOT / 'tables'
FIG_DIR.mkdir(parents=True, exist_ok=True)
TAB_DIR.mkdir(parents=True, exist_ok=True)

DATASETS = {
    'wikipedia_september_2007': RESULTS_ROOT / 'wikipedia_september_2007',
    'wiki2018': RESULTS_ROOT / 'wiki2018',
}

FEATURE_SET = 'A2'
RUN_PREFIX = '001'

plt.rcParams.update({
    'figure.dpi': 140,
    'savefig.dpi': 300,
    'font.family': 'serif',
    'font.size': 10,
    'axes.titlesize': 11,
    'axes.labelsize': 10,
    'legend.fontsize': 9,
})
```

```python
def load_summary_json(path: Path):
    with path.open() as f:
        return json.load(f)


def list_sizes(root: Path, pattern: str):
    sizes = set()
    for p in root.glob(pattern):
        try:
            sizes.add(int(p.stem.split('_')[-1]))
        except Exception:
            pass
    return sorted(sizes)


def load_il_summary(root: Path, variant: str, size: int):
    p = root / f"{RUN_PREFIX}_summary_ilnse_{FEATURE_SET}_{variant}_{size}.json"
    return load_summary_json(p)
```

```python
# Build ablation dataframe
variants = [
    ('guard_no_guard_NB', 'No-Guard'),
    ('guard_precision_only', 'Precision-only'),
    ('guard_quality_only', 'Quality-only'),
    ('guard_full_NB', 'Full-Guard'),
]

rows = []
for dataset, root in DATASETS.items():
    for variant, label in variants:
        sizes = list_sizes(root, f"{RUN_PREFIX}_summary_ilnse_{FEATURE_SET}_{variant}_*.json")
        for size in sizes:
            s = load_il_summary(root, variant, size)
            rows.append({
                'dataset': dataset,
                'variant': variant,
                'label': label,
                'cache_size': size,
                'hit_ratio': s.get('hit_ratio')
            })

ablation = pd.DataFrame(rows)
ablation.head()
```

```python
# Figure 5: Ablation HR vs cache size
for dataset in DATASETS.keys():
    data = ablation[ablation['dataset'] == dataset]
    fig, ax = plt.subplots(figsize=(6.8, 4.0))
    for variant, label in variants:
        g = data[data['variant'] == variant].sort_values('cache_size')
        ax.plot(g['cache_size'], g['hit_ratio'], marker='o', label=label)
    ax.set_xlabel('Cache size (objects)')
    ax.set_ylabel('Hit ratio')
    ax.set_title(f'Ablation HR vs Cache Size ({dataset})')
    ax.legend(loc='best', frameon=False, ncol=2)
    fig.tight_layout()
    fig.savefig(FIG_DIR / f'cont2_ablation_hr_{dataset}.png')
    fig.savefig(FIG_DIR / f'cont2_ablation_hr_{dataset}.pdf')
    plt.show()
    plt.close(fig)
```

```python
# Figure 6: Delta HR (Full - variant)
for dataset in DATASETS.keys():
    data = ablation[ablation['dataset'] == dataset]
    full = data[data['variant'] == 'guard_full_NB'][['cache_size','hit_ratio']].rename(columns={'hit_ratio':'full'})

    fig, ax = plt.subplots(figsize=(6.8, 4.0))
    for variant, label in variants:
        if variant == 'guard_full_NB':
            continue
        g = data[data['variant'] == variant][['cache_size','hit_ratio']].rename(columns={'hit_ratio':'other'})
        m = full.merge(g, on='cache_size', how='inner')
        m['delta'] = m['full'] - m['other']
        m = m.sort_values('cache_size')
        ax.plot(m['cache_size'], m['delta'], marker='o', label=f"Full - {label}")

    ax.axhline(0.0, color='black', linewidth=0.8)
    ax.set_xlabel('Cache size (objects)')
    ax.set_ylabel('ΔHR')
    ax.set_title(f'Delta HR (Full - Variant) ({dataset})')
    ax.legend(loc='best', frameon=False)
    fig.tight_layout()
    fig.savefig(FIG_DIR / f'cont2_ablation_delta_hr_{dataset}.png')
    fig.savefig(FIG_DIR / f'cont2_ablation_delta_hr_{dataset}.pdf')
    plt.show()
    plt.close(fig)
```

```python
# Table: Average delta HR and win-rate summary
rows = []
for dataset, root in DATASETS.items():
    sizes = list_sizes(root, f"{RUN_PREFIX}_summary_ilnse_{FEATURE_SET}_guard_full_NB_*.json")
    for variant, label in variants:
        if variant == 'guard_full_NB':
            continue
        deltas = []
        wins = ties = losses = 0
        for size in sizes:
            s_full = load_il_summary(root, 'guard_full_NB', size)
            s_var = load_il_summary(root, variant, size)
            if s_full.get('hit_ratio') is None or s_var.get('hit_ratio') is None:
                continue
            d = s_full['hit_ratio'] - s_var['hit_ratio']
            deltas.append(d)
            if abs(d) < 1e-6:
                ties += 1
            elif d > 0:
                wins += 1
            else:
                losses += 1
        rows.append({
            'dataset': dataset,
            'variant': label,
            'avg_delta_hr': float(np.mean(deltas)) if deltas else float('nan'),
            'win': wins,
            'tie': ties,
            'lose': losses,
        })

ablation_summary = pd.DataFrame(rows)
ablation_summary.to_csv(TAB_DIR / 'cont2_ablation_summary.csv', index=False)
ablation_summary.to_latex(TAB_DIR / 'cont2_ablation_summary.tex', index=False, float_format='%.6f')

ablation_summary
```

```python
# Optional: mechanism panel (time series) for one mid-size cache
MECH_VARIANT = 'guard_full_NB'
MECH_WINDOW = 7

for dataset, root in DATASETS.items():
    sizes = list_sizes(root, f"{RUN_PREFIX}_summary_ilnse_{FEATURE_SET}_{MECH_VARIANT}_*.json")
    if not sizes:
        continue
    size = sizes[len(sizes)//2]
    jsonl_path = root / f"{RUN_PREFIX}_ilnse_{FEATURE_SET}_{MECH_VARIANT}_{size}.jsonl"
    if not jsonl_path.exists():
        continue

    rows = []
    with jsonl_path.open() as f:
        for line in f:
            row = json.loads(line)
            if row.get('phase') != 'cache':
                continue
            rows.append(row)

    def smooth(x, w=MECH_WINDOW):
        x = np.asarray(x, dtype=float)
        out = np.full_like(x, np.nan, dtype=float)
        for i in range(len(x)):
            j0 = max(0, i - w + 1)
            out[i] = np.nanmean(x[j0:i+1])
        return out

    miss_c = np.array([r.get('miss_candidates') or 0 for r in rows], dtype=float)
    adm = np.array([r.get('admit_applied') or 0 for r in rows], dtype=float)
    adm_rate = np.divide(adm, np.maximum(1.0, miss_c))

    precision_eff = np.array([r.get('admission_precision_eff') if r.get('admission_precision_eff') is not None else np.nan for r in rows], dtype=float)
    quality_mult = np.array([r.get('quality_mult', np.nan) for r in rows], dtype=float)
    spread = np.array([r.get('score_spread_ema') if r.get('score_spread_ema') is not None else r.get('score_spread', np.nan) for r in rows], dtype=float)
    gate = np.array([1.0 if r.get('score_gate_applied') else 0.0 for r in rows], dtype=float)

    adm_rate_s = smooth(adm_rate)
    precision_s = smooth(precision_eff)
    quality_s = smooth(quality_mult)
    spread_s = smooth(spread)
    gate_s = smooth(gate)

    fig, axes = plt.subplots(3, 1, figsize=(6.6, 7.2), sharex=True)

    axes[0].plot(adm_rate_s, color='tab:blue', label='Admission rate')
    axes[0].set_ylabel('Admission rate')
    axes[0].set_title(f'Guard Mechanism (size={size}, {dataset})')
    axes[0].legend(loc='best', frameon=False)

    axes[1].plot(precision_s, color='tab:orange', label='Precision (eff)')
    axes[1].set_ylabel('Precision (eff)')
    axes[1].legend(loc='best', frameon=False)

    axes[2].plot(quality_s, color='tab:green', label='Quality mult')
    axes[2].plot(spread_s, color='tab:red', alpha=0.7, label='Score spread')
    ax2 = axes[2].twinx()
    ax2.plot(gate_s, color='tab:gray', label='Gate applied')
    axes[2].set_ylabel('Quality / Spread')
    ax2.set_ylabel('Gate applied')
    axes[2].set_xlabel('Cache slot')

    lines = axes[2].get_lines() + ax2.get_lines()
    labels = [l.get_label() for l in lines]
    axes[2].legend(lines, labels, loc='best', frameon=False)

    fig.tight_layout()
    fig.savefig(FIG_DIR / f'cont2_mechanism_{dataset}.png')
    fig.savefig(FIG_DIR / f'cont2_mechanism_{dataset}.pdf')
    plt.show()
    plt.close(fig)
```

```python
# Option A: multi-panel (1x2) delta HR ablation figure
fig, axes = plt.subplots(1, 2, figsize=(7.2, 3.4), sharey=True)

for idx, (ax, dataset) in enumerate(zip(axes, DATASETS.keys())):
    data = ablation[ablation['dataset'] == dataset]
    full = data[data['variant'] == 'guard_full_NB'][['cache_size','hit_ratio']].rename(columns={'hit_ratio':'full'})
    for variant, label in variants:
        if variant == 'guard_full_NB':
            continue
        g = data[data['variant'] == variant][['cache_size','hit_ratio']].rename(columns={'hit_ratio':'other'})
        m = full.merge(g, on='cache_size', how='inner')
        m['delta'] = m['full'] - m['other']
        m = m.sort_values('cache_size')
        ax.plot(m['cache_size'], m['delta'], marker='o', label=f"Full - {label}")

    ax.axhline(0.0, color='black', linewidth=0.8)
    ax.set_title(dataset)
    # ax.set_title(f"({chr(97+idx)})")
    ax.set_xlabel('Cache size (objects)')

axes[0].set_ylabel('ΔHR (Full − Variant)')
handles, labels = axes[0].get_legend_handles_labels()
if handles:
    fig.legend(handles, labels, loc='lower center', ncol=3, frameon=False)

# fig.suptitle('Ablation ΔHR (Contribution 2)', y=0.98)
fig.tight_layout(rect=[0, 0.12, 1, 0.93])
fig.savefig(FIG_DIR / 'cont2_ablation_delta_panel.png')
fig.savefig(FIG_DIR / 'cont2_ablation_delta_panel.pdf')
plt.show()
plt.close(fig)

```
