# Evaluation Matrices (Section VI.C / VII)


```python
from __future__ import annotations
import subprocess
from pathlib import Path

DATASET = 'wiki2018'  # or 'wikipedia_september_2007'
MODELS = ['ilnse_A2', 'ilnse_xu']  # update with your method names
BASELINE = 'ilnse_xu'  # baseline for gain table (optional)
OUT_DIR = Path('results/paper_tables')

cmd = [
    'python', 'scripts/eval_matrices.py',
    '--dataset', DATASET,
    '--out-dir', str(OUT_DIR),
    '--baseline', BASELINE,
    '--models', *MODELS,
]
print('Running:', ' '.join(cmd))
subprocess.check_call(cmd)
```

```python
# Preview generated tables
for name in [
    f'table_hr_{DATASET}.md',
    f'table_gain_{DATASET}.md',
    f'table_control_{DATASET}.md',
    f'table_stability_{DATASET}.md',
    f'table_corr_{DATASET}.md',
]:
    path = OUT_DIR / name
    if path.exists():
        print('
===', name, '===')
        print(path.read_text())
```
